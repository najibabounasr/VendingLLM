```python
import pandas as pd
import matplotlib.pyplot as plt

# Load
df = pd.read_csv("../vending_machine_sales.csv")

# Convert dates
df["TransDate"] = pd.to_datetime(df["TransDate"], errors="coerce")

# Daily sales per product (long format)
sales_per_product = (
    df.groupby(["Product", "TransDate"])
      .size()
      .reset_index(name="SalesCount")
)

# Pivot so each product is its own column
product_sales_wide = sales_per_product.pivot(
    index="TransDate", columns="Product", values="SalesCount"
).fillna(0).reset_index()

# Optional: calculate % growth for each product
product_pct_growth = product_sales_wide.set_index("TransDate").pct_change() * 100
product_pct_growth = product_pct_growth.reset_index()

# Plot example: pick one product column
example_product = product_sales_wide.columns[1]  # first product after date
plt.figure(figsize=(10,4))
plt.plot(product_sales_wide["TransDate"], product_sales_wide[example_product], marker="o")
plt.title(f"Daily Sales  Product {example_product}")
plt.xlabel("Date")
plt.ylabel("Units Sold")
plt.grid(True)
plt.show()

display(product_sales_wide.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load
df = pd.read_csv("../vending_machine_sales.csv")

# Convert dates
df["TransDate"] = pd.to_datetime(df["TransDate"], errors="coerce")

# Count daily sales per product
sales_per_product = (
    df.groupby(["TransDate", "Product"])
      .size()
      .reset_index(name="SalesCount")
)

# Pivot so each product is its own column (features)
all_products_df = sales_per_product.pivot(
    index="TransDate", columns="Product", values="SalesCount"
).fillna(0).reset_index()

# Example plot: total daily sales across all products
all_products_df["TotalSales"] = all_products_df.drop(columns="TransDate").sum(axis=1)

plt.figure(figsize=(10,4))
plt.plot(all_products_df["TransDate"], all_products_df["TotalSales"], marker="o")
plt.title("Total Daily Sales (All Products)")
plt.xlabel("Date")
plt.ylabel("Units Sold")
plt.grid(True)
plt.show()

display(all_products_df.head())
```

## Daily Sales:

```python

# Example: pick a specific product column by name
product_name = input("Enter a product name: ")   # replace with the actual product name in your data

plt.figure(figsize=(10,4))
plt.plot(all_products_df["TransDate"], all_products_df[product_name], marker="o")
plt.title(f"Daily Sales  {product_name}")
plt.xlabel("Date")
plt.ylabel("Units Sold")
plt.grid(True)
plt.show()
```

## Cumulative Sales:

```python
product_name = input("Enter a product name: ")  

# cumulative sum over time
all_products_df[f"{product_name}_Cumulative"] = all_products_df[product_name].cumsum()

# plot
plt.figure(figsize=(10,4))
plt.plot(all_products_df["TransDate"], all_products_df[f"{product_name}_Cumulative"], marker="o")
plt.title(f"Cumulative Sales  {product_name}")
plt.xlabel("Date")
plt.ylabel("Cumulative Units Sold")
plt.grid(True)
plt.show()
```
